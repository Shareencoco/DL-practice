MLP, IRIS

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, LearningCurveDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
X,y=load_iris(return_X_y=True)

X=StandardScaler().fit_transform(X)

X_tr,X_te,y_tr,y_te=train_test_split(X,y)
#baseline
m1=MLPClassifier(hidden_layer_sizes=(10,),
                 max_iter=1000,
                 random_state=0)
#l1
m2=MLPClassifier(hidden_layer_sizes=(10,),
                 max_iter=1000,
                 random_state=0,
                 alpha=0.01)
#early stop
m3=MLPClassifier(hidden_layer_sizes=(10,),
                 max_iter=1000,
                 random_state=0,
                 early_stopping=True,
                 validation_fraction=0.2)
#l1
m4=SGDClassifier(loss='log_loss',
                 penalty='l1',
                 max_iter=1000,
                 random_state=0)

models=[m1,m2,m3,m4]
for model in models:
  model.fit(X_tr,y_tr)
  LearningCurveDisplay.from_estimator(model,X_tr,y_tr,cv=5)
  plt.title(model)
  plt.show()
```
