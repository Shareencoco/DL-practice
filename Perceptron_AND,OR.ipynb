```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#LOAD DATA
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y_or=np.array([[0],[1],[1],[1]])
y_and=np.array([[0],[0],[0],[1]])

#OR
model_or = Sequential([
    Dense(1,activation='sigmoid',input_shape=(2,))
])
model_or.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model_or.fit(X,y_or,
             epochs=5,
             verbose=None)
sample = np.array([[1,0]])
predicted = (model_or.predict(sample)[0][0]>0.5).astype(int)
print(sample)
print("=================================OR============================")
print("OR PREDICTED:",predicted)
print("OR ACTUAL: 1")
(loss,accuracy) = model_or.evaluate(X,y_or)
print("acc:",accuracy)

for i,layer in enumerate(model_or.layers):
  w,b=layer.get_weights()
  print("weights\n",w)
  print("bias\n",b)

#AND
model_and = Sequential([
    Dense(1,activation='sigmoid',input_shape=(2,))
])
model_and.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model_and.fit(X,y_and,
              epochs=5,
              verbose=None)
predicted = (model_and.predict(sample)[0][0]>0.5).astype(int)
print("================================AND============================")
print("AND PREDICTED:",predicted)
print("AND ACTUAL: 0")
(loss,accuracy) = model_and.evaluate(X,y_and)
print("acc:",accuracy)


for i,layer in enumerate(model_and.layers):
  w,b=layer.get_weights()
  print("weights\n",w)
  print("bias\n",b)
```
